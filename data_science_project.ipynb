```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix 
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('./downloads/dataset.csv')
df.head(10)
```

```python
a = []
for i in df['sourceIPAddress']:
    if i == '139.198.18.205':
        a.append(1)
    else:
        a.append(0)
```

```python
df['_time'] = pd.to_datetime(df['_time'], unit='s')
df['_indextime'] = pd.to_datetime(df['_indextime'], unit='s')
df['malicious'] = a
df.head()
```

```python
df = df.sort_values('_time')
df.head()
```

```python
df_aggr = df[['_time','sourceIPAddress']]
df_aggr.set_index('_time', inplace=True, drop=True)
df_aggr.head()
```

```python
df_aggr_pv = df_aggr.pivot_table(index='_time',columns = 'sourceIPAddress', aggfunc='size', fill_value=0)
df_aggr_pv = df_aggr_pv.resample("10T").sum()
df_aggr_pv.head()
```

```python
error_codes = df['errorCode'].unique().tolist()
region = df['region'].unique().tolist()
region
```

```python
df_main = df[['_time','errorCode','sourceIPAddress']]
df_main.set_index('_time',inplace = True, drop= True)
```

```python
le = preprocessing.LabelEncoder()
le.fit(['success', 'NoSuchLifecycleConfiguration', 'NoSuchCORSConfiguration'
 ,'NoSuchEntityException', 'NoSuchTagSet', 'AccessDenied',
 'Client.UnauthorizedOperation', 'Client.InstanceLimitExceeded',
 'Client.Unsupported', 'Server.InsufficientInstanceCapacity',
 'NoSuchBucketPolicy', 'ReplicationConfigurationNotFoundError',
 'NoSuchWebsiteConfiguration'])
b = le.transform(df['errorCode'])
le.fit(['us-east-1',
 'us-west-1',
 'us-west-2',
 'ap-northeast-2',
 'sa-east-1',
 'ap-southeast-1',
 'ca-central-1',
 'ap-northeast-1',
 'ap-southeast-2',
 'us-east-2',
 'eu-west-1',
 'ap-south-1',
 'eu-west-3',
 'eu-west-2',
 'eu-central-1'])
region2 = le.transform(df['region'])
```

```python
df_aggr_pv = pd.pivot_table(df_aggr_pv, index = ['sourceIPAddress','_time'], aggfunc = 'mean').reset_index()
df_aggr_pv.head()
```

```python
df_temp = pd.DataFrame()
df_temp['errorCode'] = b
df_temp['malicious'] = df['malicious']
df_temp['sourceIPAddress'] = df['sourceIPAddress']
df_temp['region'] = region2
```

```python
# Initialize empty dictionary
error_codes_dfs = {}

# Create a dataframe for each unique errorCode and store it in
# our empty error_codes_dfs so we can call the unique df by the errorCode name
for code in error_codes:
    error_codes_dfs[code] = df_main[df_main['errorCode'] == code]

# To validate, let's get the shape of each df and
# see if the sum is the same as our original df
total_rows = 0
for code in error_codes_dfs:
    total_rows += error_codes_dfs[code].shape[0]
```

```python
features = ['region','errorCode']
y=[i for i in df_temp['malicious']]
x = []
for index,row in df_temp.iterrows():
    x.append([row['errorCode'],row['region']])
x

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 100)

clf = LogisticRegression()
            
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
scores = cross_val_score(clf, x, y, cv=10)
print(scores)
```
